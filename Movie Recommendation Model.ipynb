```python
import pandas as pd
```

```python
df = pd.read_csv("movie_dataset.csv")
```

```python
features = ['genres', 'keywords', 'overview', 'cast', 'crew', 'director']
```

```python
df = df[features + ['title']]
```

```python
for feature in features:
    df[feature] = df[feature].fillna('')
```

```python
def combine_features(row):
    return ' '.join([row[feature] for feature in features])

df['combined_features'] = df.apply(combine_features, axis=1)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer(stop_words='english')
feature_matrix = vectorizer.fit_transform(df['combined_features'])
```

```python
from sklearn.metrics.pairwise import cosine_similarity

similarity = cosine_similarity(feature_matrix)
```

```python
def recommend(movie_title):
    if movie_title not in df['title'].values:
        return "Movie not found"

    index = df[df['title'] == movie_title].index[0]
    scores = list(enumerate(similarity[index]))
    
    # Sort by similarity score
    sorted_scores = sorted(scores, key=lambda x: x[1], reverse=True)
    
    # Get top 5 recommendations (excluding itself)
    recommendations = []
    for i in sorted_scores[1:6]:
        recommendations.append(df.iloc[i[0]]['title'])
    
    return recommendations
```

```python
recommend("John Carter")
```
